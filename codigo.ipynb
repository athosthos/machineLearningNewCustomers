# Importando as bibliotecas

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

# Importando a tabela

```python
df = pd.read_csv("clientes.csv")
display(df.info())
display(df.head(5))
```

# Criando colunas de ID para as colunas de texto

Isso tem que ser feito porque a IA só consegue trabalhar com números

```python
codificador_profissao = LabelEncoder()
codificador_mix = LabelEncoder()
codificador_comportamento = LabelEncoder()

df["profissao"] = codificador_profissao.fit_transform(df["profissao"])
df["mix_credito"] = codificador_mix.fit_transform(df["mix_credito"])
df["comportamento_pagamento"] = codificador_comportamento.fit_transform(df["comportamento_pagamento"])

df.head(5)
```

# Criando o Machine Learning

Quanto mais dados tiver para treinar a máquina, melhor ela vai ser.

É dessa forma que a IA aprende a fazer previsões.

Ela entende padrões

x = dados a serem usados para prever

y = dados que queremos prever

```python
# dado que queremos prever
y = df["score_credito"]

#removendo dados que não vamos usar para prever, para manter apenas os dados que vamos usar para prever.
x = df.drop(columns=["score_credito", "id_cliente"])
```

Vamos separar uma quantidade dos dados para treinar e outros dados para comparar o treinamento.

Ou seja, se temos 100 mil dados, vamos treinar ela com 60 mil dados e os outros 40 mil dados, vamos comparar o resultado da tabela com o treinamento dela.

```python
# tem que ser sempre nessa ordem
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

Criando um modelo de previsão

- Importar a IA
- Criar a IA
- Treinar a IA

Modelos de IA: RamdomForst, Nearest Neighbors

```python
modelo_arvore = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#treinando os dois modelos
modelo_arvore.fit(x_train, y_train)
modelo_knn.fit(x_train, y_train)
```

Fazendo as previsões com as IA usando os dados de teste

```python
previsao_arvore = modelo_arvore.predict(x_test)
previsao_knn = modelo_knn.predict(x_test)
```

Comparando os modelos de IA com os dados de teste

```python
print(accuracy_score(y_test, previsao_arvore))
print(accuracy_score(y_test, previsao_knn))
print(previsao_arvore)
```

O melhor modelo é o da árvore de decisão

Se a nota de precisão do modelo está bom o suficiente ou não, vai depender do seu objetivo

Como melhor a nota:
- GridSearch
- Ter mais dados
- Testar outros parâmetros (com mais colunas, menos colunas, juntando colunas)
